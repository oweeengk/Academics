Below is the current visualization of the downloaded dataset **iris.csv** from the website. Data taken from https://www.kaggle.com/datasets/uciml/iris?resource=download

**Outliers**
    
    Let's look for possible outliers and fix them if ever

```python
import numpy
numeric_col = ['SepalLengthCm','SepalWidthCm','PetalLengthCm', 'PetalWidthCm']
categorical_col = ['Species']
df.boxplot(numeric_col)
```

There seems to be outliers only in SepalWidthCm.

Using IQR, we can follow the below approach to replace the outliers with a NULL value:

- Calculate the first and third quartile (Q1 and Q3).
- Further, evaluate the interquartile range, IQR = Q3-Q1.
- Estimate the lower bound, the lower bound = Q1*1.5
- Estimate the upper bound, upper bound = Q3*1.5
- Replace the data points that lie outside of the lower and the upper bound with a NULL value.

```python
for x in ['SepalWidthCm']:
    q75,q25 = numpy.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = numpy.nan
    df.loc[df[x] > max,x] = numpy.nan
```

Having replaced the outliers with nan, let us now check the sum of null values or missing values using the below code:

```python
df.isnull().sum()
```

```python
df = df.dropna(axis = 0)
df.isnull().sum()
```

Thus, all outliers have been treated for better presentation.
